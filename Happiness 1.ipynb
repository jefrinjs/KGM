Hypothesis testing to  happiness dataset analysis will give project more statistical depth and  analytical skills.

```python
#correlation
from scipy.stats import pearsonr

corr, p_value = pearsonr(happy['Generosity'], happy['Happiness.score'])
print("Correlation:", corr)
print("P-value:", p_value)
```

INSIGHTS:

--> The correlation = 0.16 indicates a weak positive relationship between Generosity and Happiness Score.

--> P-value ≈ 0.054 is just above 0.05, which means the result is not quite significant

--> This means we cannot confidently say that generosity affects happiness in this dataset

```python
#Test if the mean GDP per capita of Asia differs from Europe
from statsmodels.stats.weightstats import ztest

# Select GDP per capita for Asia and Europe
asia_gdp = happy[happy['Continent'] == 'Asia']['Economical rate(gdp)']
europe_gdp = happy[happy['Continent'] == 'Europe']['Economical rate(gdp)']

# Perform two-sample z-test
z_stat, p_value = ztest(asia_gdp, europe_gdp)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

INSIGHTS:

--> Z-statistic = -3.86 and p-value ≈ 0.00011 indicate that the difference in mean GDP per capita between Asia and Europe is statistically significant.

--> This means that European countries have a significantly higher GDP per capita than Asian countries on average.

--> This shows that Europe and Asia have different levels of wealth, which can affect things like how happy people are, their health, and the freedoms they enjoy



```python

from scipy.stats import ttest_ind

# Select happiness scores for Europe and Asia
europe_scores = happy[happy['Continent'] == 'Europe']['Happiness.score']
asia_scores = happy[happy['Continent'] == 'Asia']['Happiness.score']

# Perform independent t-test
t_stat, p_value = ttest_ind(europe_scores, asia_scores)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

INSIGHTS:

--> The t-statistic = 4.40 and p-value ≈ 0.00003 indicate that the difference in mean happiness scores between Europe and Asia is statistically significant.

--> This means that countries in Europe have a significantly higher average happiness score compared to countries in Asia

--> This means that, on average, countries in Europe are happier than countries in Asia.


```python
#comparing mean of happiness score among all continents
from scipy.stats import f_oneway

# Get happiness scores for each continent
europe_scores = happy[happy['Continent'] == 'Europe']['Happiness.score']
asia_scores = happy[happy['Continent'] == 'Asia']['Happiness.score']
africa_scores = happy[happy['Continent'] == 'Africa']['Happiness.score']
oceania_scores = happy[happy['Continent'] == 'Oceania']['Happiness.score']
south_america_scores = happy[happy['Continent'] == 'South America']['Happiness.score']
north_america_scores = happy[happy['Continent'] == 'North America']['Happiness.score']

# Perform one-way ANOVA
f_stat, p_value = f_oneway(europe_scores, asia_scores, africa_scores, oceania_scores, south_america_scores, north_america_scores)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

INSIGHTS:

--> There is a significant difference in average happiness scores between continents.

--> Some continents, like Europe and North America, have higher average happiness, while others, like Africa, have lower average happiness.

--> A Freedom and Governance: Countries with more personal freedom and lower perceived corruption tend to be more happier

-->This shows that where a country is located can strongly influence its citizens’ happiness.
