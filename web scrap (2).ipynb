```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import time
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument('--headless')  # For headless mode in Colab231=-
options.add_argument('--no-sandbox')=12
options.add_argument('--disable-dev-shm-usage')

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
url = "https://www.flipkart.com/food-nutrition/confectioneries-sweets/chocolates/~gourmet-foods-/pr?sid=7jv,iyp,s4n&p[]=facets.serviceability%5B%5D%3Dtrue&otracker=categorytree&otracker=nmenu_sub_Sports%2C%20Books%20%26%20More_0_Chocolates"
driver.get(url)
```

```python
choc_names =[] # empty lists for storing name, price and discount
choc_prices =[]
choc_disc = []
choc_sp = []
choc_rating = []
```

```python
num_pages = 60# number of pages to be scrapped
```

```python
from selenium import webdriver

driver = webdriver.Chrome()  # or Firefox, Edge, etc.
driver.get("https://www.flipkart.com/food-nutrition/confectioneries-sweets/chocolates/~gourmet-foods-/pr?sid=7jv,iyp,s4n&p[]=facets.serviceability%5B%5D%3Dtrue&otracker=categorytree&otracker=nmenu_sub_Sports%2C%20Books%20%26%20More_0_Chocolates")

content = driver.page_source
```

```python
for page in range(num_pages):
    time.sleep(5)  # reduce waiting time; use WebDriverWait if needed
    content = driver.page_source
    soup = BeautifulSoup(content, 'html.parser')  # reading the page source
    chocolate = soup.find_all('div', class_="cPHDOP col-12-12")  # all products
    
    for i in chocolate:
        # Chocolate Name
        names = i.find('a', class_="wjcEIp")
        if names:
            choc_names.append(names.get_text(strip=True))
        else:
            choc_names.append('N/A')

        # MRP (original price)
        MRP = i.find('div', class_="yRaY8j")
        if MRP:
            choc_prices.append(MRP.get_text(strip=True))
        else:
            choc_prices.append("N/A")

        # Discount
        disc = i.find('div', class_="UkUFwK")
        if disc:
            choc_disc.append(disc.get_text(strip=True))
        else:
            choc_disc.append("N/A")

        # Selling Price
        selling_price = i.find('div', class_="Nx9bqj")
        if selling_price:
            choc_sp.append(selling_price.get_text(strip=True))
        else:
            choc_sp.append("N/A")

        # Rating
        rating = i.find('div', class_="XQDdHH")
        if rating:
            choc_rating.append(rating.get_text(strip=True))
        else:
            choc_rating.append("N/A")

    try:
        nxt_button = driver.find_element("xpath", '//a[@class="_9QVEpD"]')  # locate next button
        nxt_button.click()
    except Exception as e:
        print(e)
        break

driver.quit()

# ✅ Build DataFrame
data = {
    "choc_name": choc_names,
    "MRP": choc_prices,
    "Discount": choc_disc,
    "Selling_Price": choc_sp,
    "Rating": choc_rating
}

df = pd.DataFrame(data)
print(df)

df.to_csv("chocolate.csv", index=False)
```
